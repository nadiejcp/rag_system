```python
import requests
from bs4 import BeautifulSoup

url = "https://subslikescript.com/movies"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36"
}

resp = requests.get(url, headers=headers, timeout=30)
print("STATUS:", resp.status_code)
print("FINAL URL:", resp.url)
print("TEXT START:")
print(resp.text[:1000])

# Intento de inspección básica del DOM
soup = BeautifulSoup(resp.text, "html.parser")
print("TITLE:", soup.title.get_text(strip=True) if soup.title else None)

# Ver algunos links del documento
links = []
for a in soup.select("a[href]")[:20]:
    links.append((a.get_text(" ", strip=True), a["href"]))
print("FIRST LINKS:")
for text, href in links:
    print(text, "=>", href)
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import re

url = "https://subslikescript.com/movies"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36"
}

html = requests.get(url, headers=headers, timeout=30).text
soup = BeautifulSoup(html, "html.parser")

# 1) Buscar links que parezcan películas
candidates = []
for a in soup.select("a[href]"):
    href = a["href"].strip()
    text = " ".join(a.get_text(" ", strip=True).split())
    if any(k in href.lower() for k in ["/movie", "/transcript", "/script"]) or any(k in text.lower() for k in ["movie", "transcript", "script"]):
        candidates.append((text, href))

# deduplicar conservando orden
seen = set()
unique_candidates = []
for item in candidates:
    if item not in seen:
        unique_candidates.append(item)
        seen.add(item)

print("CANDIDATE LINKS:", len(unique_candidates))
for text, href in unique_candidates[:80]:
    print(f"{text!r} -> {href}")

# 2) Mostrar el bloque de contenedor principal
main = soup.find("main")
print("\nMAIN TAG:", bool(main))
if main:
    print(main.get_text(" ", strip=True)[:2000])

# 3) Buscar paginación
print("\nPAGINATION LINKS:")
for a in soup.select("a[href]"):
    text = " ".join(a.get_text(" ", strip=True).split())
    href = a["href"]
    if text.isdigit() or text.lower() in ["next", "prev", "previous"] or "page" in href.lower():
        print(f"{text!r} -> {href}")

# 4) Inspeccionar clases más comunes del HTML
from collections import Counter
classes = Counter()
for tag in soup.find_all(True):
    cls = tag.get("class")
    if cls:
        classes[" ".join(cls)] += 1

print("\nMOST COMMON CLASSES:")
for cls, n in classes.most_common(20):
    print(n, cls)
```
